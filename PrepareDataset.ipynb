```python
import numpy as np
from scipy.io import wavfile
```

```python
def SplitAudio (fileName,talk,pause,subjID):
    sampleRate, data = wavfile.read(fileName)
    wavLen = len(data)
    idx = np.arange(1,150*(talk+pause)*sampleRate,(talk+pause)*sampleRate)
    for i in range(len(idx)):
        start = idx[i]
        end  = start + talk*sampleRate
        track = data[start:end+1]
        trackName = 'S' + '%02d' % (subjID) + '_Train_' + 'Sent' + '%03d' % (i+1) + '.wav'
        wavfile.write(trackName,sampleRate,track)
        

def GenerateTranscript (fileName,subjID):
    c = 0;
    for i in open (fileName):
        transcriptFileName = 'S' + '%02d' % (subjID) + '_Train_' + 'Sent' + '%03d' % (c+1) + '.txt'
        txt = i.split('\n')[0]
        
        if not i in ['\n', '\r\n']:
            txt = i.split('.')[1]
            f = open(transcriptFileName, "w")
            f.write(txt)
            f.close()
            c+=1
            
def RenameEMG (directory,extension,subjID):
    from stat import S_ISREG, ST_CTIME, ST_MODE
    import os, sys, time
    
    # path to the directory (relative or absolute)
    dirpath = directory

    # get all entries in the directory w/ stats
    entries = (os.path.join(dirpath, fn) for fn in os.listdir(dirpath) if fn.endswith(extension))
    entries = ((os.stat(path), path) for path in entries)

    # leave only regular files, insert creation date
    entries = ((stat[ST_CTIME], path)
               for stat, path in entries if S_ISREG(stat[ST_MODE]))
    #NOTE: on Windows `ST_CTIME` is a creation date 
    #  but on Unix it could be something else
    #NOTE: use `ST_MTIME` to sort by a modification date
    c = 0
    for cdate, path in sorted(entries):
        
        newFileName = 'S' + '%02d' % (subjID) + '_Train_' + 'Sent' + '%03d' % (c+1) + extension
        #os.rename(os.path.basename(path), os.path.basename(path).split('.')[0] + 'Hello' + extension)
        os.rename(os.path.basename(path), newFileName)
        c+=1
        print time.ctime(cdate), os.path.basename(path)
```
